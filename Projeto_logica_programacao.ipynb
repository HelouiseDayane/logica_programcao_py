Explicação do código:

A função knn() recebe como parâmetros o conjunto de treinamento, os rótulos do conjunto de treinamento, o ponto a ser classificado e o número de vizinhos.


---


1.   É implementado pela variável k, que recebe o valor do número de vizinhos.

2.   É implementado pela função np.linalg.norm(), que calcula a distância entre o ponto a ser classificado e cada ponto do conjunto de treinamento. A função np.argsort() retorna os índices dos k vizinhos mais próximos do ponto a ser classificado, ordenados em ordem crescente de distância.
3. É implementado de acordo com o tipo de problema. Para problemas de regressão, a classe do ponto a ser classificado é a média dos valores dos k vizinhos mais próximos. Para problemas de classificação, a classe do ponto a ser classificado é a moda dos valores dos k vizinhos mais próximos.
Parâmetros do algoritmo KNN:

K: Número de vizinhos. Quanto maior o valor de K, mais suave será a fronteira de decisão do modelo.
Função de distância: Função que calcula a distância entre dois pontos. A função np.linalg.norm() é uma função genérica que pode ser usada para calcular diferentes tipos de distâncias.
Vantagens e desvantagens do algoritmo KNN:



---



Vantagens:
 * Simples de entender e implementar.
 * Não requer pré-processamento dos dados.
 * Pode ser usado para problemas de regressão e classificação.


---
Desvantagens:

 * Pode ser ineficiente para grandes conjuntos de dados.
 * Pode ser sensível a outliers.

```python
# Função para calcular a distância Euclidiana entre dois vetores
def euclidean_distance(vec1, vec2):
    # Calcula a soma dos quadrados das diferenças entre os elementos dos dois vetores
    squared_diff = sum((x - y) ** 2 for x, y in zip(vec1, vec2))
    # Tira a raiz quadrada da soma para obter a distância Euclidiana
    return squared_diff ** 0.5

# Função para classificar um cliente não classificado com base no algoritmo k-NN
def classify_knn(data, no_class, k):
    classifications = []  # Lista para armazenar as classificações dos clientes não classificados
    for unclassified in no_class:
        unclassified_data = unclassified[2]  # Dados do cliente não classificado
        distances = []  # Lista para armazenar as distâncias entre o cliente não classificado e os clientes classificados
        for customer in data:  # Itera sobre os clientes classificados
            customer_data = customer[2]  # Dados do cliente classificado
            dist = euclidean_distance(unclassified_data, customer_data)  # Calcula a distância Euclidiana
            distances.append((customer, dist))  # Armazena o cliente classificado e a distância
        distances.sort(key=lambda x: x[1])  # Ordena os clientes classificados com base na distância
        nearest_neighbors = distances[:k]  # Seleciona os k vizinhos mais próximos
        classes = [neighbor[0][1] for neighbor in nearest_neighbors]  # Obtém as classes dos vizinhos mais próximos
        # Verifica se é um problema de classificação ou regressão
        if all(c.isdigit() for c in classes):
            # Problema de Regressão: Calcula a média das classes dos vizinhos
            avg_class = sum(float(c) for c in classes) / k
            classifications.append((unclassified[0], avg_class))  # Armazena a classificação
        else:
            # Problema de Classificação: Calcula a moda das classes dos vizinhos
            mode_class = max(set(classes), key=classes.count)
            classifications.append((unclassified[0], mode_class))  # Armazena a classificação
    return classifications  # Retorna as classificações dos clientes não classificados

# Define o valor de k
k = 5

# Classifica os clientes não classificados
classifications = classify_knn(data, no_class, k)

# Imprime as classificações
for classification in classifications:
    print(f"Cliente CPF: {classification[0]}, Classificação NPS: {classification[1]}")
```

```python
# Função para calcular a distância Euclidiana entre dois vetores
def euclidean_distance(vec1, vec2):
    # Inicializa uma variável para armazenar a soma dos quadrados das diferenças
    sum_of_squares = 0

    # Itera através dos elementos dos dois vetores ao mesmo tempo
    for x, y in zip(vec1, vec2):
        # Calcula a diferença entre os elementos dos vetores
        difference = x - y
        # Soma o quadrado da diferença à variável de soma
        sum_of_squares += difference ** 2

    # Calcula a raiz quadrada da soma dos quadrados
    euclidean_dist = sum_of_squares ** 0.5

    return euclidean_dist

# Função de comparação para ordenar com base nas distâncias
def compare_distances(item):
    # Extrai o valor da distância (o segundo elemento da tupla)
    return item[1]

# Função para classificar um cliente não classificado com base no algoritmo k-NN
def classify_knn(data, no_class, k):
    # Inicializa uma lista para armazenar as classificações
    classifications = []

    # Itera através dos clientes não classificados
    for unclassified in no_class:
        unclassified_data = unclassified[2]
        distances = []

        # Itera através dos clientes classificados para calcular as distâncias
        for customer in data:
            customer_data = customer[2]
            dist = euclidean_distance(unclassified_data, customer_data)
            distances.append((customer, dist))

        # Ordena as distâncias com base na função de comparação
        distances.sort(key=compare_distances)

        # Pega os k vizinhos mais próximos
        nearest_neighbors = distances[:k]

        # Cria uma lista com as classes dos vizinhos mais próximos
        classes = [neighbor[0][1] for neighbor in nearest_neighbors]

        # Verifica se é um problema de classificação ou regressão
        if all(c.isdigit() for c in classes):
            # Problema de Regressão
            avg_class = sum(float(c) for c in classes) / k
            classifications.append((unclassified[0], avg_class))
        else:
            # Problema de Classificação
            # Encontra a classe mais frequente
            mode_class = max(set(classes), key=classes.count)
            classifications.append((unclassified[0], mode_class))

    return classifications

# Define o valor de k
k = 5

# Classifica os clientes não classificados
classifications = classify_knn(data, no_class, k)

# Imprime as classificações
for classification in classifications:
    print(f"Cliente CPF: {classification[0]}, Classificação NPS: {classification[1]}")
```
